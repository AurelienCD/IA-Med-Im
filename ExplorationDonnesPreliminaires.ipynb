### Librairies :

```python
import pandas as pad
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
custom_palette = [sns.xkcd_rgb["windows blue"], sns.xkcd_rgb["pale red"], sns.xkcd_rgb["medium green"], "orange", "blue","yellow", "purple", "deeppink", "brown", "teal", "black"] 
sns.set_palette(custom_palette)
```

### Exploration de la base des données :

```python
# Lecture du fichier Excel 

df = pad.read_excel('Database_analyse.xlsx', header = 0, sheet_name = 'Total')
print('La base de données totale comporte', len(df), 'patients.')

df_homme = pad.read_excel('Database_analyse.xlsx', header = 0, sheet_name = 'H')
print('La base de données homme comporte', len(df_homme), 'patients.')

df_femme = pad.read_excel('Database_analyse.xlsx', header = 0, sheet_name = 'F')
print('La base de données femme comporte', len(df_femme), 'patients.')
```

```python
def ExplorationDonneesPreliminaires_diagramme(df, df_femme, df_homme, colonne):  
    
    fig, axs = plt.subplots(1, 3, figsize=(20, 10)) 
    
    data       = (df.value_counts()*100)/len(df)
    data_femme = (df_femme.value_counts()*100)/len(df)
    data_homme = (df_homme.value_counts()*100)/len(df)

    axs[0].pie(x=data, labels=data.index, colors=custom_palette, autopct='%1.2f%%')
    axs[0].set_title('Répartition globale')

    # Diagramme circulaire pour le DataFrame femme
    axs[1].pie(x=data_femme, labels=data_femme.index, colors=custom_palette, autopct='%1.2f%%')
    axs[1].set_title('Répartition femmes')

    # Diagramme circulaire pour le DataFrame homme
    axs[2].pie(x=data_homme, labels=data_homme.index, colors=custom_palette, autopct='%1.2f%%')
    axs[2].set_title('Répartition hommes')

    plt.show()
    fig.savefig(f'{colonne}_diagramme.png')
```

```python
def ExplorationDonneesPreliminaires_histplot_boxplot(df, df_femme, df_homme, colonne):
    
    ## Valeur minimale, maximale, moyenne et écart-type des données
    print('Valeur minimale :', round(df.min(), 2))
    print('Valeur maximale :', round(df.max(), 2))
    print('Valeur moyenne :', round(df.mean(), 2))
    print('Valeur écart-type :', round(df.std(), 2))
        
    fig, axs = plt.subplots(2, 3, figsize=(20, 10))  
    
    ## Histogramme de distribution des données
    sns.histplot(data=df, x=df, kde=False, ax=axs[0,0], color = custom_palette[0]) 
    axs[0,0].set_title('Globale')
    
    sns.histplot(data=df_femme, x=df_femme, kde=False, ax=axs[0,1], color = custom_palette[1]) 
    axs[0,1].set_title('Femmes')
    
    sns.histplot(data=df_homme, x=df_homme, kde=False, ax=axs[0,2], color = custom_palette[2]) 
    axs[0,2].set_title('Hommes')
        
    ## Boxplot des données
    sns.boxplot(data=df, ax=axs[1,0], color = custom_palette[0])
    sns.stripplot(data=df, dodge=True, ax=axs[1,0], color = custom_palette[0])
    axs[1,0].set_title('Global')
    
    sns.boxplot(data=df_femme, ax=axs[1,1], color = custom_palette[1])
    sns.stripplot(data=df_femme, dodge=True, ax=axs[1,1], color = custom_palette[1])
    axs[1,1].set_title('Femmes')
    
    sns.boxplot(data=df_homme, ax=axs[1,2], color = custom_palette[2])
    sns.stripplot(data=df_homme, dodge=True, ax=axs[1,2], color = custom_palette[2])
    axs[1,2].set_title('Hommes')
        
    plt.tight_layout()
    plt.show()  
    fig.savefig(f'{colonne}_histo_boxplot.png')
```
